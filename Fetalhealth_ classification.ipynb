```python
train_cols=df.columns[~df.columns.isin(['fetal_health'])]
train_cols
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
scaler = StandardScaler()
X= scaler.fit_transform(df[train_cols])
X = pd.DataFrame(X, columns=train_cols)
X.head()
```

```python
from sklearn.feature_selection import SelectKBest, chi2,SelectFromModel
from sklearn import linear_model
from sklearn.linear_model import LassoCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import sklearn
for v in [linear_model.Lasso(),LogisticRegression(penalty="l2"),RandomForestClassifier(n_estimators=100)]:
    kb = SelectFromModel(v) 
    kb.fit(X, df['fetal_health'])
    t = X.columns[kb.get_support(True)]
    print(t.values)

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
rfe_selector = RFE(estimator=LogisticRegression(max_iter=200), n_features_to_select=12, step=10, verbose=5)
rfe_selector.fit(X, df['fetal_health'])
rfe_support = rfe_selector.get_support()
rfe_feature = X.loc[:,rfe_support].columns.tolist()
print(str(len(rfe_feature)), 'selected features')
print(rfe_feature)
```

```python
df_x = df[train_cols]
X, test, y, y_test = sklearn.model_selection.train_test_split(df_x, df['fetal_health'], stratify=df['fetal_health'], 
                                                              test_size=.1)
pd.Series(y_test).value_counts()
```

```python
X.shape
```

```python
X = scaler.fit_transform(X)
test = scaler.transform(test)
X
```

```python
y.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping
```

```python
def build_model():  
    model = models.Sequential()
    model.add(layers.Dense(20, activation='relu', input_shape=(X.shape[1],)))
    model.add(layers.Dense(20, activation='relu'))
    model.add(layers.Dense(3, activation='softmax'))
    
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=[tf.metrics.Recall(), 'accuracy'])      
    return model
model = build_model()
model.summary()
```

```python

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
model = KerasClassifier(build_fn=build_model, epochs=80, batch_size=8, verbose=1)
modelfit = model.fit(X, y)
```

```python
plt.plot(modelfit.history['accuracy'])
 
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch') 
plt.legend(['train', 'validation'], loc='upper left') 
plt.show()   
```
